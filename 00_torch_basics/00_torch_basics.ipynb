## matrix multiplication

1. we have normal element vise multiplication
2. we have dot product, or matrix multiplication

torch.matmul ~= @ or torch.mm

```python
# tensors
t1 = torch.tensor([[1,2,3],
                   [4,5,6],
                   [2,3,4],
                   [5,7,9]])

t2 = torch.tensor([[8,9,5],
                   [8,3,6],
                   [2,3,4],
                   [5,7,9]])
```

```python
t1.T
```

```python
t2
```

```python
t1.T.shape, t2.shape
```

```python
#matrix multiplication
torch.matmul(t1.T,t2)
```

```python
torch.mm(t1.T, t2) ## torch.mm work on 2d without broadcasting
```

```python
# for 1d
t1[0].dot(t2[0])
```

```python
%%time
t1*t2
```

```python
%%time 
torch.matmul(t1.T,t2)
```

#matmul vs mm

```python
mat1 = torch.rand(4,3)
mat2 = torch.rand(3,4)

# matmul only work when RxC ka shape == CxR ka shape
print(torch.matmul(mat1, mat2))

mat1 = torch.rand(4,3)
mat2 = torch.rand(3,6)

print(mat1)
print(mat2)

# mm works without broadcasting so RxC ka shape != CxR ka shape me bhi kaam karega
print(torch.mm(mat1, mat2))

```

## Reshaping , Stacking, Squeezing and unsqueezing tensors

* reshape - reshape input dim
* view - share same memory of og tensor, just show in diff dims.
* stacking - stack, vstack, hstack.
* squeeze - remove all `1` dimensions from a tensor
* unsqueeze - add a `1` dimemsion to a target tensor
* permute - return a view of the input with dimensions permute (swapped) in a certain way.

```python
x = torch.arange(2.,12.)
```

```python
x, x.shape
```

```python
x.view(2,5)
```

```python
x.reshape(2,5)
```

```python
# stacking

torch.stack((x,x,x),dim=1) # horizontal stack
```

```python
torch.hstack((x,x,x))
```

```python
torch.stack((x,x,x),dim=0) # vertical stack
```

## Pytorch tensor & numpy

* numpy -> tensor `torch.from_numpy(ndarray)`
* tensor -> numpy `torch.Tensor.numpy()`

```python
# from numpy to tensor
import torch
import numpy as np

arr = np.array([1,2,3,4,5])
trch = torch.from_numpy(arr)

trch+=1 # use var = var + 1 , not var+=1

arr, trch
```

```python
arr+=1
trch, arr
```

```python

# it will create it in int64 coz default numpy is int64
```

```python
# from tensor to numpy
tsor = torch.ones(10)
nump = tsor.numpy()

tsor = tsor + 1
nump, tsor # share same memory space
```
