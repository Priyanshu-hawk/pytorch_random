```python
import os
IMAGE_PATH = "../dataset/fgvc-aircraft-2013b/data/images"

all_imgs = []

with open("../dataset/fgvc-aircraft-2013b/data/images_test.txt", "r") as f:
    for i in f.readlines():
        all_imgs.append(os.path.join(IMAGE_PATH, i.strip()+".jpg"))

print(os.path.exists(all_imgs[25]))
Image.open(all_imgs[25])
```

```python
image_annote = []
with open("../dataset/fgvc-aircraft-2013b/data/images_box.txt", "r") as f:
    for i in f.readlines():
        image_annote.append(i.strip().split(" "))
```

```python
img_name = {}

t = ["test", "train", "val"]

for tit in t:
    with open("../dataset/fgvc-aircraft-2013b/data/images_variant_{}.txt".format(tit)) as f:
        for name in f.readlines():
            # if len(name) > 0:
                # print(name.strip().split(" "))
                img_class, img_obj = name.strip().split(" ")[0],name.strip().split(" ")[1:] 
                img_name[img_class] = img_obj

print(len(img_name))
```

```python
import cv2
import random
import matplotlib.pyplot as plt

rand_id = random.randint(0, len(image_annote))
img = cv2.imread(os.path.join(IMAGE_PATH, image_annote[rand_id][0]+".jpg"))
# print(rand_id, (image_annote[rand_id][1], image_annote[rand_id][2]))
plt.imshow(cv2.rectangle(img, \
  (int(image_annote[rand_id][1]), int(image_annote[rand_id][2])), \
  (int(image_annote[rand_id][3]), int(image_annote[rand_id][4])), (255,0,0)))
print(img_name[image_annote[rand_id][0]])
```
